```python
import pandas as pd
import numpy as np
import datetime as dt

df = pd.read_csv("../data/processed/online_retail_clean.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

df.head()
```

```python
orders = (
    df.groupby(["CustomerID", "InvoiceNo"], as_index=False)
      .agg(
          InvoiceDate=("InvoiceDate", "min"),
          Revenue=("Revenue", "sum")
      )
)

orders.head()
```

```python
snapshot_date = orders["InvoiceDate"].max() + dt.timedelta(days=1)
snapshot_date
```

```python
# rfm table
rfm = orders.groupby("CustomerID").agg(
    Recency=("InvoiceDate", lambda x: (snapshot_date - x.max()).days),
    Frequency=("InvoiceNo", "nunique"),
    Monetary=("Revenue", "sum")
)

rfm.head()
```

```python
rfm.describe()
rfm.isna().sum()
```

```python
rfm_transformed = rfm.copy()
rfm_transformed["Frequency"] = np.log1p(rfm["Frequency"])
rfm_transformed["Monetary"] = np.log1p(rfm["Monetary"])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_transformed)
```

```python
# choosing no of clusters
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(2, 9):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(rfm_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 9), inertia, marker="o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for k Selection")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_profile = rfm.groupby("Cluster").agg(
    Customers=("Recency", "count"),
    AvgRecency=("Recency", "mean"),
    AvgFrequency=("Frequency", "mean"),
    AvgMonetary=("Monetary", "mean")
)

cluster_profile
```

```python
# naming the segments
segment_map = {
    0: "High-Value Loyal",
    1: "Recent Big Spenders",
    2: "At-Risk Customers",
    3: "Low Engagement"
}

rfm["Segment"] = rfm["Cluster"].map(segment_map)
rfm.head()
```

```python
#contribution of revenue by segmants
segment_revenue = (
    rfm.groupby("Segment")["Monetary"]
       .sum()
       .sort_values(ascending=False)
)

(segment_revenue / rfm["Monetary"].sum()).apply(lambda x: f"{x:.2%}")
```

```python
segment_size = rfm["Segment"].value_counts(normalize=True)
segment_size.apply(lambda x: f"{x:.2%}")
```

```python
# checking if high value customers stay valuable
cutoff = orders["InvoiceDate"].quantile(0.7)

early = orders[orders["InvoiceDate"] <= cutoff]
late = orders[orders["InvoiceDate"] > cutoff]
```

```python
# early rfm period
snapshot_early = cutoff + pd.Timedelta(days=1)

rfm_early = early.groupby("CustomerID").agg(
    Recency=("InvoiceDate", lambda x: (snapshot_early - x.max()).days),
    Frequency=("InvoiceNo", "nunique"),
    Monetary=("Revenue", "sum")
)
```

```python
#camparison with future revenue
late_revenue = late.groupby("CustomerID")["Revenue"].sum()

validation = rfm_early.join(late_revenue, how="left")
validation.columns = ["Recency", "Frequency", "Monetary", "FutureRevenue"]

validation.groupby(pd.qcut(validation["Monetary"], 4))["FutureRevenue"].mean()
```

```python
rfm_out = rfm[["Recency", "Frequency", "Monetary", "Segment"]].copy()
rfm_out.to_csv("../data/processed/rfm_segments.csv")

print("Saved rfm_segments.csv")
```

Customers were segmented using Recency, Frequency and Monetary value at the order level to avoid line-item bias. After stabilizing skewed distributions and clustering scaled features, distinct customer segments emerged, including high-value loyal customers, recent big spenders, at-risk customers and low-engagement customers. 

Revenue analysis confirmed that a small, high-value segment accounts for a disproportionate share of total revenue, validating the business importance of targeted customer strategies.
