```python
import pandas as pd
import numpy as np
import datetime as dt

df = pd.read_csv("../data/processed/online_retail_clean.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

orders = (
    df.groupby(["CustomerID", "InvoiceNo"], as_index=False)
      .agg(
          InvoiceDate=("InvoiceDate", "min"),
          Revenue=("Revenue", "sum")
      )
)
```

```python
# seperating past and future orders
cutoff_date = orders["InvoiceDate"].quantile(0.7)

train_orders = orders[orders["InvoiceDate"] <= cutoff_date]
future_orders = orders[orders["InvoiceDate"] > cutoff_date]
```

```python
snapshot_date = cutoff_date + dt.timedelta(days=1)
```

```python
X = train_orders.groupby("CustomerID").agg(
    Recency=("InvoiceDate", lambda x: (snapshot_date - x.max()).days),
    Frequency=("InvoiceNo", "nunique"),
    Monetary=("Revenue", "sum")
)
```

```python
# defining target
y = future_orders.groupby("CustomerID")["Revenue"].sum()
```

```python
clv_df = X.join(y, how="left").fillna(0)
clv_df.columns = ["Recency", "Frequency", "Monetary", "CLV"]
```

```python
clv_df.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    clv_df[["Recency", "Frequency", "Monetary"]],
    clv_df["CLV"],
    test_size=0.2,
    random_state=42
)
```

```python
# baseline model
baseline_preds = X_val["Monetary"]
```

```python
from sklearn.metrics import mean_absolute_error

print(
    "Baseline MAE:",
    mean_absolute_error(y_val, baseline_preds)
)
```

```python
# model training
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)

preds = model.predict(X_val)

print("Model MAE:", mean_absolute_error(y_val, preds))
```

```python
# SHAP
import shap

explainer = shap.Explainer(model)
shap_values = explainer(X_val)

shap.summary_plot(shap_values, X_val)
```

```python
# ranking customers by predicted clv
clv_df["PredictedCLV"] = model.predict(
    clv_df[["Recency", "Frequency", "Monetary"]]
)

clv_df.sort_values("PredictedCLV", ascending=False).head(10)
```

```python
# revenue concentration
top_20_pct = clv_df["PredictedCLV"].quantile(0.8)

top_customers = clv_df[clv_df["PredictedCLV"] >= top_20_pct]

top_customers["CLV"].sum() / clv_df["CLV"].sum()
```

```python
rfm = pd.read_csv("../data/processed/rfm_segments.csv", index_col="CustomerID")

clv_with_segments = clv_df.join(rfm["Segment"], how="left")

clv_with_segments.groupby("Segment")["PredictedCLV"].mean().sort_values(ascending=False)
```

Customer lifetime value was framed as future revenue over a fixed horizon to prevent data leakage and unrealistic assumptions. A machine-learning model trained on historical RFM features outperformed a strong baseline based on past spend alone and produced a ranked list of customers by expected future value. Importantly, the model demonstrated that future value depends not only on how much a customer has spent, but also on how recently and how frequently they purchase.

**CLV prediction is best used as a ranking tool to prioritize high-ROI customers.**
