## Data Cleaning

```python
# droping missing values
df = df.dropna(subset=["CustomerID"])
```

```python
# remove cancellations (negative quantity or invoice starting with C)
df = df[(df["Quantity"] > 0) & (~df["InvoiceNo"].astype(str).str.startswith("C"))]
```

```python
# create revenue
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
# convert date
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

## Exploratory Data Analysis

```python
# countries that drive the most revenue
df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
```

```python
# daily revenue trend
daily_revenue = df.groupby(df["InvoiceDate"].dt.date)["Revenue"].sum()
daily_revenue.plot(figsize=(12,4), title="Daily Revenue Trend")
```

```python
# revenue distribution
customer_revenue = df.groupby("CustomerID")["Revenue"].sum()
customer_revenue.describe(percentiles=[0.5, 0.8, 0.9, 0.95, 0.99])
```

```python
customer_revenue.sort_values(ascending=False).head(10)
```

```python
# checking typical order value vs outliers and quantity distribution per order
order_value = df.groupby("InvoiceNo")["Revenue"].sum()
order_value.describe()
```

```python
order_value.plot(kind="hist", bins=50, figsize=(8,4), title="Order Value Distribution")
```

```python
# checking one time buyers vs repeaters
orders_per_customer = df.groupby("CustomerID")["InvoiceNo"].nunique()
orders_per_customer.value_counts(normalize=True).head(10)
```

```python
# days between purchases
df_sorted = df.sort_values(["CustomerID", "InvoiceDate"])

df_sorted["PrevPurchase"] = df_sorted.groupby("CustomerID")["InvoiceDate"].shift(1)
df_sorted["DaysBetween"] = (df_sorted["InvoiceDate"] - df_sorted["PrevPurchase"]).dt.days
```

```python
df_sorted["DaysBetween"].describe()
```

```python
# monthly trends
df["Month"] = df["InvoiceDate"].dt.to_period("M")
df.groupby("Month")["Revenue"].sum().plot(figsize=(10,4))
```

```python
# weekly behavior
df["DayOfWeek"] = df["InvoiceDate"].dt.day_name()
df.groupby("DayOfWeek")["Revenue"].mean()
```

```python
# product analysis
product_revenue = df.groupby("StockCode")["Revenue"].sum()
product_revenue.describe(percentiles=[0.8, 0.9, 0.99])
```

```python
# revenue by country
df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head()
```

```python
df[["Quantity", "UnitPrice", "Revenue"]].corr()
```

The raw UCI Online Retail dataset required significant preprocessing to ensure analytical validity, including removing cancellations, handling missing customer identifiers and correcting invalid quantities and prices. 

Exploratory analysis revealed strong revenue skew, with a small percentage of customers and products accounting for most sales, as well as clear seasonality in purchasing behavior. 

These findings motivated the use of customer-level aggregation, lifecycle analysis, and robust modeling techniques in subsequent steps.
